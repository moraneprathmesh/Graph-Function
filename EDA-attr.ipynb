```python
dat[['Gender','Attrition']].groupby(['Gender']).mean()
#attrition rate is higher in males
```

```python
dat[['Gender','Attrition']].groupby(['Gender']).mean().plot(kind='bar', figsize=(4,4), fontsize=12,rot=45)
plt.xlabel('Gender',fontsize=12)
plt.ylabel('Attrition rate',fontsize=12)
plt.title('Attrition Rate With Resp To Gender',fontsize=14)
plt.show()
#higher attrition in males visible from graph
```

```python
dat.groupby('EducationField').size().plot(kind='pie', autopct='%.2f')
plt.title('EducationField breakup',fontsize=14)
plt.show()
dat.groupby(['EducationField'])['Attrition'].mean().plot(kind='bar', figsize=(4,4), fontsize=12)
plt.xlabel('EducationField',fontsize=12)
plt.ylabel('Attrition rate',fontsize=12)
plt.title('Attrition With Resp To EducationField',fontsize=14)
plt.show()
#while there are very few people of hr and technical degree but there attrition rate is quiet high may be these people are not satisfied with work in company
```

```python
dat.groupby('Department').size().plot(kind='pie', autopct='%.2f')
plt.title('Department',fontsize=14)
plt.show()
dat.groupby(['Department'])['Attrition'].mean().plot(kind='bar', figsize=(4,4), fontsize=12)
plt.xlabel('Department',fontsize=12)
plt.ylabel('Attrition rate',fontsize=12)
plt.title('Attrition With Resp To Department',fontsize=14)
plt.show()
#similarly in this also attrtion is high in dept sales and hr 
```

```python
dat.groupby('BusinessTravel').size().plot(kind='pie', autopct='%.2f')
plt.title('BusinessTravel breakup',fontsize=14)
plt.show()
dat.groupby(['BusinessTravel'])['Attrition'].mean().plot(kind='bar', figsize=(4,4), fontsize=12,rot=45)
plt.xlabel('BusinessTravel',fontsize=12)
plt.ylabel('Attrition rate',fontsize=12)
plt.title('Attrition With Resp To BusinessTravel',fontsize=14)
plt.show()
#travel frequently(which is 20% of total employees) is having very high attrition rate which is 25%
```

```python
dat.groupby('OverTime').size().plot(kind='pie', autopct='%.2f')
plt.title('OverTime breakup',fontsize=14)
plt.show()
dat.groupby(['OverTime'])['Attrition'].mean().plot(kind='bar', figsize=(4,4), fontsize=12)
plt.xlabel('OverTime',fontsize=12)
plt.ylabel('Attrition rate',fontsize=12)
plt.title('Attrition With Resp To OverTime',fontsize=14)
plt.show()
# attrition rate of people which are made to do overtime is very high around 30 % which is aroung 60% percent of total employess left
```
